```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the dataset
# Assuming the dataset is in a CSV file named 'data.csv'
data = pd.read_csv('Data_set.csv')

# Extract the column containing reasons for investment choices
# Assuming the reasons column is named 'Investment_Reason'
reasons_data = data['What are your savings objectives?']

# Function to clean and process text data
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = ''.join(e for e in text if e.isalnum() or e.isspace())
    return text

# Apply preprocessing to reasons data
cleaned_reasons = reasons_data.apply(preprocess_text)

# Tokenize the reasons and count the frequency of each word
all_words = ' '.join(cleaned_reasons).split()
word_counts = Counter(all_words)

# Identify the most common words (excluding common stop words)
common_stop_words = {'the', 'and', 'for', 'to', 'a', 'in', 'of', 'is', 'with', 'that', 'it', 'as', 'on', 'are', 'this', 'by', 'an', 'be', 'at', 'or', 'from', 'we', 'will', 'not'}
common_words = {word: count for word, count in word_counts.items() if word not in common_stop_words}

# Display the most common reasons
most_common_reasons = Counter(common_words).most_common(10)
print("Most common reasons for investment choices:")
for reason, count in most_common_reasons:
    print(f"{reason}: {count}")

# Generate a word cloud to visualize the common reasons
wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=100).generate_from_frequencies(common_words)

# Plot the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Common Reasons for Investment Choices')
plt.show()
```
