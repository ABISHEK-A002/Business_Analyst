```python
import pandas as pd
```

```python
data=pd.read_csv('Data_set.csv')
```

```python
data.head()
```

```python
gender=data['gender']
age=data['age']
df = data[['gender', 'age']]
```

```python
count= gender.value_counts()
```

```python
count
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=count.index, y=count.values, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
count.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis', len(count)))
plt.ylabel('')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=age, y=gender, hue=gender, palette='viridis', s=100, alpha=0.6, edgecolor='w', linewidth=0.5)
plt.xlabel('Age')
plt.ylabel('Gender')
plt.title('Scatter Plot of Gender Distribution by Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(gender, bins=len(count), discrete=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution - Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=gender, y=age, data=df, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Gender - Box Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='gender', y='age', data=df, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Gender - Violin Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='gender', y='age', data=df)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Gender - Swarm Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(x='gender', y='age', data=df, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Gender - Strip Plot')
plt.show()
```

```python
g = sns.FacetGrid(df, col="gender", height=6, aspect=1)
g.map(sns.histplot, "age")
g.set_axis_labels("age", "Count")
g.set_titles("Gender")
g.fig.suptitle('Age Distribution by Gender - Facet Grid', y=1.02)
plt.show()
```
